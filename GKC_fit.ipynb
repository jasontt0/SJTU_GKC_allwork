```python
# 导入必要的库
import numpy as np

# 给定的数据点
x = np.array([10.555, 12.916, 15.322, 17.625, 19.910, 26.628])
y = np.array([10, 20, 30, 40, 50, 80])

# 计算各项
n = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x ** 2)

# 计算斜率 m 和截距 b
m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x ** 2)
b = (sum_y - m * sum_x) / n

# 输出结果
print(f"拟合的一次函数为: y = {m:.4f}x + {b:.4f}")

# 可视化拟合结果
import matplotlib.pyplot as plt

# 生成拟合直线的数据点
x_fit = np.linspace(min(x), max(x), 100)
y_fit = m * x_fit + b

# 绘制原始数据点和拟合直线
plt.scatter(x, y, color='blue', label='原始数据点')
plt.plot(x_fit, y_fit, color='red', label='拟合直线')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('一次函数拟合')
plt.show()
```

### 以下使用scikit-learn实现线性回归

```python
# 导入必要的库
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 给定的数据点
x = np.array([48.56, 80.40, 23.60]).reshape(-1, 1)  # 将 x 转换为二维数组，因为 scikit-learn 需要这样的输入
y = np.array([50.19, 81.45, 25.98])

# 创建并训练线性回归模型
model = LinearRegression()
model.fit(x, y)

# 获取拟合的斜率 m 和截距 b
m = model.coef_[0]
b = model.intercept_

# 输出结果
print(f"拟合的一次函数为: y = {m:.4f}x + {b:.4f}")

# 生成拟合直线的数据点
x_fit = np.linspace(min(x), max(x), 100).reshape(-1, 1)
y_fit = model.predict(x_fit)

# 绘制原始数据点和拟合直线
plt.scatter(x, y, color='blue', label="原始数据点")
plt.plot(x_fit, y_fit, color='red', label='拟合直线')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('一次函数拟合')
plt.show()
```
