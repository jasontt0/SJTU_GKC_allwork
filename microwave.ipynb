```python
import matplotlib.pyplot as plt

# 创建一个新的图形
plt.figure(figsize=(10, 6))

# 设置x轴和y轴的范围
plt.xlim(0, 180)
plt.ylim(-60, 0)

# 添加网格
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# 设置x轴和y轴的标签
plt.xlabel('X Axis (0 to 180)')
plt.ylabel('Y Axis (-60 to 0)')

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib

# 设置中文字体
matplotlib.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
matplotlib.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 创建一个新的图形
plt.figure(figsize=(10, 6))

# 设置x轴和y轴的范围
plt.xlim(0, 180)
plt.ylim(-60, 0)

# 添加网格
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# 设置x轴和y轴的标签
plt.xlabel('X 轴 (0 到 180)')
plt.ylabel('Y 轴 (-60 到 0)')

# 设置标题
plt.title('网格图示例')

# 显示图形
plt.show()
```
